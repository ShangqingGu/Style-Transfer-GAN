```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
stat_location_0 = "/datasets/home/home-03/84/484/yhc001/Style-Transfer-GAN/assets/Exp0-photo2vangogh/statistics/"
index = [x for x in range(0,2832,48)]
index.append(2831)
```

```python
D_loss_0 = []
G_loss_0 = []
adv_0 = []
cycle_0 = []
identity_0 = []
for i in range(0,11):
    file = open(stat_location_0+str(i))
    lines = file.readlines()
    file.close()
    for x in index:
        category = lines[x].strip().split(' ')[1:]
        for p in range(len(category)):
            category[p] = category[p].split(':')[1]
        D_loss_0.append(float(category[0]))
        G_loss_0.append(float(category[1]))
        adv_0.append(float(category[2]))
        cycle_0.append(float(category[3]))
        identity_0.append(float(category[4]))
```

```python
plot_index = np.array(index)/2831
```

```python
test = plot_index
for x in range(1,11):
    test = np.append(test,plot_index+1*x+.00001)
```

```python
test 
```

```python
plt.figure(figsize=(20,20))
plt.plot(test,D_loss_0, label="Discriminator Loss")
plt.plot(test,G_loss_0, label="Generator Loss")
plt.plot(test,adv_0, label="Adversarial Loss")
plt.plot(test,cycle_0, label="Cycle Consistency Loss")
plt.plot(test,identity_0, label="Identity Loss")
plt.xlabel("Epoch")
plt.legend()
plt.title("Experiment 0")
#plt.ylim([0,6])
plt.xticks([x for x in range(12)])
plt.savefig('exp0.png')
```
